## Analiza danych źródłowych

Dane dotyczą **lotów krajowych wykonanych w 2015 roku przez amerykańskie linie lotnicze**. Zestaw został przygotowany na podstawie comiesięcznego raportu "Air Travel Consumer Report" wykonywanego przez Biuro Statystyki Transportu Departamentu Transportu Stanów Zjednoczonych (DOT).

Zestaw zawiera 3 pliki csv: *airlines*, *airports* oraz *flights*. 

Tabele *airlines* i *airports* to tabele wymiarów, zawierające atrybuty opisowe konkretnych linii lotniczych i lotnisk. Dane służą jedynie do grupowania i filtrowania wyników analizy.

Tabela *flights* jest tabelą faktów, zawierającą miliony rekordów pojedynczych lotów oraz kluczowe miary liczbowe (np. opóźnienie, dystans, czas kołowania), na których możemy wykonywać obliczenia i agregacje. 

## Strategia SCD

Dla tabel wymiarów używamy logiki SCD2, aby zachować historię zmian atrybutów. Celowo powielamy wiersze, aby umożliwić połączenie historycznych faktów z poprawnmyi danymi (np. nazwą lotniska/linii lotniczej), które wtedy obowiązywały. 

Do każdego rekordu dopisujemy 3 kolumny: *is_current* (True/False), *start_date* (default: 1900-01-01), *end_date* (default: null), które będą wskazywać na to, czy informacje są dalej aktualne, a jeśli nie są, to od kiedy do kiedy aktualne były. 

Dzięki temu podejściu, jeśli któryś z atrybutów się zmieni, rekord zostanie zamknięty (*is_current*=False, *end_date*=data_zmiany). Dla danej linii do tabeli zostanie dopisany nowy wiersz, zawierający aktualne informacje.

## Diagram ERD

Hurtownia danych jest modelowana w *schemacie gwiazdy*. Dzięki temu podejściu, tabele wymiarów mogą być bezpośrednio połączone z tabelą faktów, co zapewnia najwyższą wydajność. Dodatkowo prosta i intuicyjna struktura ułatwia analizę.
